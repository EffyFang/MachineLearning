```python



folder_benign_train = 'Data/train/benign'
folder_malignant_train = 'Data/train/malignant'

read = lambda imname: np.asarray(Image.open(imname).convert("RGB"))
ims_benign = [read(os.path.join(folder_benign_train, filename)) for filename in os.listdir(folder_benign_train)]
X_benign = np.array(ims_benign, dtype='uint8')
ims_malignant = [read(os.path.join(folder_malignant_train, filename)) for filename in os.listdir(folder_malignant_train)]
X_malignant = np.array(ims_malignant, dtype='uint8')
```

```python
# load vgg model

from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16

folder_benign_train = 'Data/train/benign'
folder_malignant_train = 'Data/train/malignant'

read = lambda imname: img_to_array(load_img(imname, target_size=(224, 224)))
ims_benign = [read(os.path.join(folder_benign_train, filename)) for filename in os.listdir(folder_benign_train)]

ims_malignant = [read(os.path.join(folder_malignant_train, filename)) for filename in os.listdir(folder_malignant_train)]
```

```python
y_benign = np.zeros(len(ims_benign))
y_malignant = np.ones(len(ims_malignant))
```

```python
#Merge data
X_train_ims = np.concatenate((ims_benign, ims_malignant), axis = 0)
y_train_ims = np.concatenate((y_benign, y_malignant), axis = 0)

#shuffle
np.random.seed(0)
s = np.arange(X_train_ims.shape[0])
np.random.shuffle(s)
X_train_ims = X_train_ims[s]
y_train_ims = y_train_ims[s]
X_train_ims = preprocess_input(X_train_ims)
```

```python
from keras.layers import Dense
from keras.models import Sequential

model = Sequential()
batch_size = 500
# load the model
model.add(VGG16())
model.add(Dense(units =1, activation='softmax'))

model.compile( "adam", loss='binary_crossentropy',  metrics=['accuracy'])
#model.compile(optimizer = 'sgd')

```

```python
model.summary()
```

```python
model.fit(X_train_ims, y_train_ims,epochs=3, batch_size=batch_size)
#model.save_weights('bottleneck_fc_model.h5')

# summarize the model
```
